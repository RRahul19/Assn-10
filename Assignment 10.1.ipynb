```python
import pandas as pd

Baby_Names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

babyNames=pd.DataFrame(Baby_Names)
```

```python
#Deleting Unwanted Columns

babyNames.drop(babyNames.loc[:,babyNames.columns.str.contains('^Unnamed')], axis =1, inplace = True)

babyNames.head(50)
```

```python
#Distribution of Male and Females

babyNames.Gender.value_counts()
```

```python
#top 5 most preferred names

babyNames.groupby(['Name'])['Count'].sum().nlargest(5)
```

```python
#Median name occured in dataset

babyNames[babyNames.Id == babyNames.median().Id]['Name']
```

```python
#Distribution of male and female born count by states

babyNames.groupby(['State','Gender'])['Count'].sum()
```
